```python
import cv2
import torch
import matplotlib.pyplot as plt
import numpy as np


from google.colab.patches import cv2_imshow
from PIL import Image
from numpy import asarray
```

```python
def create_filters(filter_params):
    filter_row, filter_col, num_filters = filter_params
    filters = [np.random.rand(filter_row, filter_col, 3) for _ in range(num_filters)]
    return filters
```

```python
def apply_convolution(image, filters):
    feature_maps = []
    for filter_ in filters:
        convolved_feature = [cv2.filter2D(image, -1, filter_[:, :, i]) for i in range(filter_.shape[2])]
        feature_maps.append(convolved_feature)
    return feature_maps
```

```python
def main():
    # Resmi Oku
    image = cv2.imread('doga.jpg')

    # Kullanıcıdan giriş al
    filter_dimensions = list(map(int, input("Filtre boyutlarını (row, column, filtre sayısı) girin: ").split()))
    stride_padding = list(map(int, input("Stride ve padding değerlerini girin: ").split()))

    # Filtreleri oluştur
    filters = create_filters(filter_dimensions)

    # Convolüsyon uygula
    feature_maps = apply_convolution(image, filters)

    """# Kullanıcıya feature map sayısı kadar sonuçları göster
    for idx, feature_map in enumerate(feature_maps):
      for i, convolved_feature in enumerate(feature_map):
        cv2_imshow(convolved_feature)"""
    # Her bir feature map'i ayrı ayrı göstermek için döngü
    for idx, feature_map in enumerate(feature_maps):
      print(f"Feature Map {idx + 1}")  # Hangi feature map'in gösterildiğini belirten bilgi mesajı
      for i, convolved_feature in enumerate(feature_map):
        cv2_imshow(convolved_feature)


    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
if __name__ == "__main__":
    main()
```
